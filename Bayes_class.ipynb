```python
import numpy as np
import pandas as pd
```

```python
data_train=pd.read_csv('HW5traindata.csv')
data_train
```

```python
y=data_train['y']
x=data_train.drop('y',axis=1)
```

```python
def p_y(y):
    p_y=[]
    for i in np.sort(y.unique()):
        pyi=len(y[y==i])/len(y)
        p_y.append(pyi)
    return p_y
class_probs=p_y(y)
class_probs
```

```python
sd_1=x['x1'].std()
sd_2=x['x2'].std()
sd_2
```

```python
x1=x['x1']
x2=x['x2']
r = np.corrcoef(x1, x2)
p_c = r[0,1]
```

```python
def mean_for_each_class(x,y):
    means=[]
    for i in np.sort(y.unique()):
        means_i=x[y==i].mean()
        means.append(means_i)
    return means
means = mean_for_each_class(x,y)
means
```

```python
def calculateGaussianProbability(x1, x2, u1k, u2k, st_1, st_2, p): #(µ1k, µ2k, σ1, σ2, ρ),
    expo = np.exp(-(np.power(st_2, 2)*np.power((x1-u1k), 2))+ np.power(st_1, 2)*np.power((x2-u2k), 2)- (2*p*st_1*st_2*(x1-u1k)*(x2-u2k)) / 2*(1-np.power(p,2))*st_1*st_1*st_2*st_2)
    return (1 / (2*np.pi*st_1*st_2*(np.sqrt(1-np.power(p,2))))) * expo
```

```python
def calculateClassProbabilities(x,means,st1,st2,class_probs):
    probabilities = []
    for i in range(len(class_probs)):
        numerator=1
        mean_1k=means[0][i]
        mean_2k=means[1][i]
        fxji=calculateGaussianProbability(x1=x.iloc[:,0],x2=x.iloc[:,1], u1k=mean_1k, u2k=mean_2k,st_1=sd_1,st_2=sd_2, p=p_c)
        #fxji=calculateGaussianProbability(x=x.iloc[:,j], mean=mean_ij, stdev=sd_ij)
        numerator=numerator*fxji
        numerator=numerator*class_probs[i]
        probabilities.append(numerator)
    return (probabilities/np.sum(probabilities,axis=0))
```

```python
calculateClassProbabilities(x,means,sd_1,sd_2,class_probs)
data_test = pd.read_csv('HW5testdata.csv')
x_t=data_test.drop('y',axis=1)
x_t
```

```python
def predict(x):
    probabilities=calculateClassProbabilities(x_t,means,sd_1,sd_2,class_probs)
    Class=(probabilities[1]>=0.5).astype('int')
    return(Class)
```

```python
prediction=predict(x_t)
list(zip(prediction,data_test.y))
```

```python
def compute_accuracy(prediction,test_y):
    return (1-np.sum(np.abs(prediction-test_y))/len(prediction))*100
compute_accuracy(prediction,data_test.y)
```
